Research Question #2 : Is there an association between Communication with father and Independency?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

*Codebook* 1

H4WP37.png

```python
#Manage data for variable 1
df['Communication_Father'] = df['H4WP37']
df['Communication_Father'] = df['Communication_Father'].loc[ df['Communication_Father'] !=' ']
df['Communication_Father'] = df['Communication_Father'].loc[ df['Communication_Father'] !='6']
df['Communication_Father'] = df['Communication_Father'].loc[ df['Communication_Father'] !='7']
df['Communication_Father'] = df['Communication_Father'].loc[ df['Communication_Father'] !='8']
df['Communication_Father'].replace(['1','2','3','4','5'] , ['strongly agree','agree','neither agree nor disagree','disagree','strongly disagree' ], inplace = True)
df['Communication_Father'].value_counts()
```

*Codebook* 2

H4PE38.png

```python
#Manage data for variable 2
df['Independency'] = df['H4PE38']
df['Independency'] = df['Independency'].loc[ df['Independency'] !=' ']
df['Independency'] = df['Independency'].loc[ df['Independency'] !='6']
df['Independency'] = df['Independency'].loc[ df['Independency'] !='8']
df['Independency'] = df['Independency'].loc[ df['Independency'] !='.']
df['Independency'].replace(['1','2','3','4','5'] , ['strongly agree','agree','neither agree nor disagree','disagree','strongly disagree' ], inplace = True)
df['Independency'].value_counts()
```

```python
rp.summary_cat(df['Independency'])
```

```python
#Graphs and tables or statistics
```

```python
plt.title("Other people determine most of what I can and cannot do.")
df['Independency'].value_counts().plot(kind ='pie',autopct='%1.1f%%',startangle = 320)
```

```python
plt.ylabel("Percent")
plt.title("Other people determine most of what I can and cannot do.")
df['Independency'].value_counts(normalize = True).plot(kind ='barh',color = 'Blue', width = 0.4, alpha = 0.4)
```
